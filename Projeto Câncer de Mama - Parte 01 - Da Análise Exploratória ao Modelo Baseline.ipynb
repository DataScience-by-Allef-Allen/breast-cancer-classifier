```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

features = df.drop(columns=['target'])
features_scaled = StandardScaler().fit_transform(features)

pca = PCA(n_components=2)
pca_components = pca.fit_transform(features_scaled)

pca_df = pd.DataFrame(data=pca_components, columns=['PCA1', 'PCA2'])
pca_df['target'] = df['target'].map({0: 'Benigno', 1: 'Maligno'})


sns.scatterplot(x='PCA1', y='PCA2', hue='target', data=pca_df, palette='coolwarm')
plt.title('Visualização com PCA')
plt.show()
```

## Pré-Processamento

### Corrigindo espaçamento entre palavras

```python
df.columns
```

```python
df.columns = [coluna.replace(' ', '_') for coluna in df.columns]
```

```python
df.columns
```

### Contagem de outliers

```python
import pandas as pd


outliers_por_coluna = {}


for coluna in df.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]
    outliers_por_coluna[coluna] = outliers.shape[0]

outliers_df = pd.DataFrame(list(outliers_por_coluna.items()), columns=['Coluna', 'Número de Outliers'])

print(outliers_df)
```

## Divisão , normalização ou padronização e baseline:

### Separar features (X) e alvo (y)

```python
X = df.drop(columns=['target'])
y = df['target']
```

```python
print(f"Shape do DataFrame de features: {X.shape}")
display(X)
```

### Divisão treino e teste com estratificação

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

### Normalização com MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
X_train_minmax = minmax_scaler.fit_transform(X_train)
X_test_minmax = minmax_scaler.transform(X_test)
```

### Selecionar as Top 10 Variáveis (por Importância de Random Forest)

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
importances = pd.Series(model.feature_importances_, index=X.columns)
ordered_importances = importances.sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=ordered_importances.values, y=ordered_importances.index, palette='viridis')
plt.title('Importância das Características')
plt.xlabel('Importância')
plt.ylabel('Características')
plt.show()
```

```python
top_10_variables = ordered_importances.head(10).index.tolist()

indices_top10 = [X.columns.get_loc(col) for col in top_10_variables]

X_train_top10_minmax = X_train_minmax[:, indices_top10]
X_test_top10_minmax = X_test_minmax[:, indices_top10]
```

```python
print("Top 10 variáveis selecionadas:")
print(top_10_variables)
```

### Recriar DataFrames e salvar como CSV

```python
X_train_top10_df = pd.DataFrame(X_train_top10_minmax, columns=top_10_variables)
X_test_top10_df = pd.DataFrame(X_test_top10_minmax, columns=top_10_variables)

train_df = X_train_top10_df.copy()
train_df['target'] = y_train.values

test_df = X_test_top10_df.copy()
test_df['target'] = y_test.values

train_df.to_csv('cancer_train_top10_minmax.csv', index=False)
test_df.to_csv('cancer_test_top10_minmax.csv', index=False)
```

## Baseline de Classificação:

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

train_df = pd.read_csv('cancer_train_top10_minmax.csv')
test_df = pd.read_csv('cancer_test_top10_minmax.csv')

X_train = train_df.drop(columns='target')
y_train = train_df['target']

X_test = test_df.drop(columns='target')
y_test = test_df['target']

baseline = DummyClassifier(strategy='most_frequent')
baseline.fit(X_train, y_train)

y_pred = baseline.predict(X_test)

print("Acurácia do baseline:", accuracy_score(y_test, y_pred))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred, zero_division=0))


cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Maligno', 'Benigno'], yticklabels=['Maligno', 'Benigno'])
plt.xlabel('Classe Predita')
plt.ylabel('Classe Real')
plt.title('Matriz de Confusão - Baseline')
plt.tight_layout()
plt.show()
```
