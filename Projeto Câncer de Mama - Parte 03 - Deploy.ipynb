```python
import gradio as gr
import joblib
import pandas as pd
import warnings

warnings.filterwarnings("ignore", category=UserWarning, module="gradio")

model = joblib.load('best_model_xgb_grid.pkl')

def predict_cancer_class(worst_concave_points, worst_area, worst_perimeter, worst_radius, mean_radius,
                         mean_perimeter, mean_concave_points, mean_concavity, mean_area, worst_concavity):

    input_df = pd.DataFrame([[worst_concave_points, worst_area, worst_perimeter, worst_radius, mean_radius,
                              mean_perimeter, mean_concave_points, mean_concavity, mean_area, worst_concavity]],
                            columns=['worst_concave_points', 'worst_area', 'worst_perimeter', 'worst_radius',
                                     'mean_radius', 'mean_perimeter', 'mean_concave_points', 'mean_concavity',
                                     'mean_area', 'worst_concavity'])

    prob = model.predict_proba(input_df)[0][1]
    classe = 'Benigno' if prob > 0.60 else 'Maligno'
    return f"{classe} (Confiança: {prob:.2%})"

with gr.Blocks(title="Classificador de Câncer de Mama") as interface:
    gr.Markdown("<h1 style='text-align: center;'>🎗️ Classificador de Câncer de Mama</h1>")
    gr.Markdown("<p style='text-align: center;'>Preencha os campos abaixo com os valores extraídos do exame para prever se o tumor é <b>benigno</b> ou <b>maligno</b>.</p>")

    with gr.Row():
        with gr.Column():
            worst_concave_points = gr.Number(label='Piores Pontos Côncavos', minimum=0, step=0.000001)
            worst_area = gr.Number(label='Pior Área', minimum=0, step=0.000001)
            worst_perimeter = gr.Number(label='Pior Perímetro', minimum=0, step=0.000001)
            worst_radius = gr.Number(label='Pior Raio', minimum=0, step=0.000001)
            worst_concavity = gr.Number(label='Pior Concavidade', minimum=0, step=0.000001)
        with gr.Column():
            mean_radius = gr.Number(label='Raio Médio', minimum=0, step=0.000001)
            mean_perimeter = gr.Number(label='Perímetro Médio', minimum=0, step=0.000001)
            mean_concave_points = gr.Number(label='Pontos Côncavos Médios', minimum=0, step=0.000001)
            mean_concavity = gr.Number(label='Concavidade Média', minimum=0, step=0.000001)
            mean_area = gr.Number(label='Área Média', minimum=0, step=0.000001)

    btn = gr.Button("Classificar", elem_id="classificar-btn")
    output = gr.Textbox(label="Resultado da Classificação")

    interface.css = """
    #classificar-btn {
        background-color: #FF54A7;
        color: white;
    }
    """


    def wrapper_fn(wcp, wa, wp, wr, mr, mp, mcp, mc, ma, wc):
        return predict_cancer_class(wcp, wa, wp, wr, mr, mp, mcp, mc, ma, wc)

    btn.click(fn=wrapper_fn,
              inputs=[worst_concave_points, worst_area, worst_perimeter, worst_radius, mean_radius,
                      mean_perimeter, mean_concave_points, mean_concavity, mean_area, worst_concavity],
              outputs=output)

interface.launch()
```
